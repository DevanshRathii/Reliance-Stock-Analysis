```python
import pandas as pd
import yfinance as yf
import datetime
from datetime import date, timedelta
import plotly.graph_objects as go
import plotly.express as px

#CANDLESTICK


df = pd.read_csv('/RILO - Copy (1).csv')


fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open Price'],
                high=df['High Price'],
                low=df['Low Price'],
                close=df['Close Price'])])
fig.show()
```

```python
#LINE GRAPH 

figure = px.line(df, x="Date", y="Close Price",
                 title='Reliance Stock Analysis with rangeslider') 
figure.update_xaxes(rangeslider_visible=True) 
figure.show()
```

```python
df.head()
```

```python
# Finding high variance entries

df['Modulus']=abs(df['Close Price']-df['Open Price'])
```

```python
df.head()
```

```python
df1=df.loc[df['Modulus']>110]
df1
```

```python
#LINE GRAPH 

figure = px.line(df1, x="Date", y="Modulus",
                 title='Specific Values') 
figure.update_xaxes(rangeslider_visible=True) 
figure.show()
```
